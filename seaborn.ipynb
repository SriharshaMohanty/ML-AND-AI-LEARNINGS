```python

import numpy as np 
import pandas as pd
import seaborn as sns
df = pd.read_csv("https://raw.githubusercontent.com/SriharshaMohanty/Yhills_July12_Analytics/main/titanic-training-data.csv")
df.head()
```

```python
df.dtypes
```

```python
sns.relplot(x="Name",y="Age",hue="Embarked",data=df)
```

```python
sns.catplot(x="Pclass",y="Age",data=df)
```

```python
sns.catplot(x="Pclass",y="Age",jitter=False,data=df)
```

```python
sns.catplot(x="Pclass",y="Age",hue="Sex",data=df)
```

```python
sns.catplot(x="Pclass",y="Age",kind="swarm",data=df)
```

```python
sns.catplot(x="Pclass",y="Age",kind="box",data=df)
```

```python
sns.catplot(x="Pclass",y="Age",hue="Sex",kind="violin",data=df)
```

```python
sns.catplot(x="Pclass",y="Age",hue="Sex",kind="violin",split=True,data=df)
```

```python
sns.catplot(x="Pclass",y="Age",hue="Sex",kind="point",data=df)
```
