```python
import numpy as np
import pandas as pd 
import seaborn as sns
```

```python
df = pd.read_csv("https://raw.githubusercontent.com/Premalatha-success/Datasets/main/World%20Happiness%20Report.csv")
df.head()
```

```python
df.dtypes
```

```python
sns.relplot(x="Happiness Rank",y="Economy",hue="Health",data=df)
```

```python
sns.catplot(x="Family",y="Freedom",data=df)
```

```python
sns.catplot(x="Family",y="Health",jitter=False,data=df)
```

```python
sns.catplot(x="Corruption",y="Region",kind="swarm",data=df)
```

```python
sns.catplot(x="Health",y="Region",kind="box",data=df)
```

```python
sns.catplot(x="Health",y="Region",kind="violin",data=df)
```

```python
sns.catplot(x="Happiness Rank",y="Economy",hue="Health",kind="point",data=df)
```

```python
sns.catplot(x="Job Satisfaction",y="Region",hue="Happiness Rank",kind="point",split=True,data=df)
```
