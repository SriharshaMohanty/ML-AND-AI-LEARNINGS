MATPLOTLIB

```python
import matplotlib.pyplot as plt
import numpy as np
```

```python
xpoints = np.array([1,2])
ypoints = np.array([1,2])
plt.plot(xpoints, ypoints)
plt.show()
```

```python
xpoints = np.array([1,20])
ypoints = np.array([10,200])
plt.plot(xpoints, ypoints)
plt.show()
```

```python
xpoints = np.array([1,6])
ypoints = np.array([4,5])
plt.plot(xpoints, ypoints,'*')
plt.show()
```

```python
ypoints = np.array([3,8,1,10])
plt.plot(ypoints, marker = 'o',ms=30,mec ='r' ,color = 'blue')
plt.show()
```

```python
xpoints = np.array([3,8,1,12])
ypoints = np.array([5,6,7,9])
plt.plot(xpoints,ypoints, marker = '*',ms=30,mec ='r' ,color = 'red')
plt.show()
```

```python
ypoints = np.array([3,8,1,12])
xpoints=np.array([5,6,7,9])
plt.plot(xpoints,ypoints,marker ='+',ms=100,color='red')
plt.show()
```

```python
ypoints=np.array([3,8,1])
plt.plot(ypoints,marker='o',ms=30,mec='b',color='yellow')
plt.show()
```

```python
x=np.array([80,85,90,95,100,105,110,115,120,125])
y=np.array([240,250,260,270,280,290,300,310,320,330])

plt.title("sports watch data")
plt.xlabel("Average pulse")
plt.ylabel("calorie Burnage")

plt.plot(x,y,'k')

plt.grid(color="green",linestyle="dotted")

plt.show()
```

```python
x=np.array([80,85,90,95,100,105,110,115,120,125])
y=np.array([240,250,260,270,280,290,300,310,320,330])
plt.scatter(x,y)
```

```python
x=np.array(["A","B","C","D","E","F"])
y=np.array([240,250,260,270,280,290])
plt.bar(x,y,color="orange")
plt.show()
```

```python
x=np.array(["A","B","C","D","E","F"])
y=np.array([240,250,260,270,280,290])
plt.barh(x,y,color="orange")
plt.show()
```

```python
plt.hist(x)
plt.show()
```

```python
plt.pie(y)
plt.show()
```

```python
mylabels=["apples","bananas","cherries","dates","coconut","watermelon"]
plt.pie(y,labels=mylabels,startangle=90)
plt.show()
```

```python
y=np.array([35,25,15,45,55,78])
mylabels=["apples","bananas","cherries","dates","coconut","watermelon"]
myexplode=[0,0.95,0.8,0.5,0.6,0.03]
plt.pie(y,labels = mylabels,explode = myexplode)
plt.show()
```

```python
y=np.array([35,25,15,45,55,78])
mylabels=["apples","bananas","cherries","dates","coconut","watermelon"]

plt.pie(y)
plt.legend(labels=mylabels)
plt.show()
```
